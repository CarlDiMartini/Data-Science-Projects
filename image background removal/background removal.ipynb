```python
import numpy as np
import requests
```

```python
def background_remover(picture_file):
    response = requests.post(
        'https://api.remove.bg/v1.0/removebg',
        files={'image_file': open(picture_file, 'rb')},
        data={'size': 'auto'},
        headers={'X-Api-Key': 'pZNrbEB5AzsKU7VUcML7HHdh'},
    )
    if response.status_code == requests.codes.ok:
        with open('no_background_' + str(picture_file), 'wb') as out:
            out.write(response.content)
    else:
        print("Error:", response.status_code, response.text)
    return
```

```python
background_remover('lazy dog.jpg')
```

```python
background_remover('wedding gown.jpg')
```
